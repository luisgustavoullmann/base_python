```python
# Data - Função do pandas_datareader que importa dados da web
from pandas_datareader import data as web
```

```python
# Função DataReader, busca na fonte yahoo
# dados da Proctor and Gamble, começando de 1/1/1995
# 1 - Simbulo da ação (ticker em inglês)
# 2 - Fonte dos dados
# 3 - Data de inicial da busca dos dados
PG = web.DataReader('PG', data_source='yahoo', start='1995-1-1')
PG
```

```python
# Mostrando 5 colunas de preço ajustado de 5 empresas diferentes
# Só funciona se estiverem organizados da mesma forma, com os mesmo nomes nas colunas

tickers = ["PG", "MSFT", "T", "F", "GE"]
# new_data: Pandas Data Frame de cada valor da lista tickers
new_data = pd.DataFrame()
for t in tickers:
    # t = valor de tickers em seu índice da lista
    # new_data[t]: Recebe o data frame da lista tickers em sua posição
    # De 0 até -1 da lista, retornado ["coluna que quer"]
    new_data[t] = web.DataReader(t, data_source="yahoo", start="1995-1-1")["Adj Close"]
```

```python
# Imprime na tela apenas as 5 últimas observações
new_data.tail()
```
