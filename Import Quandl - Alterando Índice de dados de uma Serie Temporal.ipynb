```python
# Ou pode usar o set_index()
# set_index() apenas exibirá os dados com o índice que você indicou
# Atribua a uma variavel a ela mesma
# ele não altera a tabela definitivamente como index_col
data_02.set_index('Date') # Date se tornou o indice!
```
